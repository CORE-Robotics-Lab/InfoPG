```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def compare_batch(data, n_agents, start_epoch, end_epoch, batches):
    fig, ax = plt.subplots(1,n_agents, figsize = (35,8), constrained_layout=True)
    fig1, ax_total = plt.subplots(figsize = (20, 10))

    for i in range(0, len(data)):
        d = data[i][0]
        label = data[i][1]
        cum_reward = np.zeros((n_agents, batches))
        cum_reward_total = np.zeros((end_epoch - start_epoch, n_agents, batches))
        
        for batch in range(0, batches):
            for agent_ix in range(0, n_agents):
                cum_reward[agent_ix, batch] = d[start_epoch][agent_ix][0][batch]
        cum_reward_total[start_epoch] = cum_reward
                
        for i in range(start_epoch+1, end_epoch):
            assert len(d[i]) == n_agents, 'Mismatch between data agent num, and n_agents: %s, %s' % (len(d[i]), n_agents)
            for batch in range(0, batches):
                for agent_ix in range(0, n_agents):
                    cum_reward[agent_ix, batch] = 0.05 * d[i][agent_ix][0][batch] + (1 - 0.05) * cum_reward[agent_ix, batch]
            cum_reward_total[i] = cum_reward
        
        mean_agent_performance = cum_reward_total.mean(axis=2)
        min_agent_performance = cum_reward_total.min(axis=2)
        max_agent_performance = cum_reward_total.max(axis=2)
        
        for agent_ix in range(0, n_agents):
            x = np.arange(start_epoch, end_epoch)
            ax[agent_ix].plot(x, mean_agent_performance[start_epoch:end_epoch, agent_ix], label = label)
            ax[agent_ix].fill_between(x, min_agent_performance[start_epoch:end_epoch, agent_ix], max_agent_performance[start_epoch:end_epoch, agent_ix], alpha=0.2)
            agent = 'Agent_%s' % (agent_ix)
            ax[agent_ix].set_title(agent)
            ax[0].set_xlabel('epochs')
            ax[0].set_ylabel('cumulative reward')

        team_performance = cum_reward_total.sum(axis=1)
        mean_team_performance = team_performance.mean(axis=1)
        min_team_performance = team_performance.min(axis=1)
        max_team_performance = team_performance.max(axis=1)

        x_axis = np.arange(start_epoch, end_epoch)
        ax_total.plot(x_axis, mean_team_performance, label=label)
        ax_total.set_ylim(-80, 25)
        ax_total.fill_between(x_axis, min_team_performance, max_team_performance, alpha=0.2)
    plt.legend()
    fig.suptitle('Agent Performance over Time', fontsize=16)
    fig1.suptitle('Team Performance over Time')
    fig.show()
```

```python
with open(os.path.join('experiments', 'pr2-ac', '2021-09-28 13_39_16', 'trajectories_data.npy'), 'rb') as f:
    data = np.load(f)
```

```python
compare_batch([[data, 'pr2-ac']], 2, 0, 999, 16)
```
